```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)
```

Creating Raw Data Directory

```python
BASE_PATH = r"D:\Dynamic Price Optimization Recommendation Engine"
RAW_DATA_PATH = os.path.join(BASE_PATH, "data", "raw")

os.makedirs(RAW_DATA_PATH, exist_ok=True)
```

Creating Products Dataset

```python
products = pd.DataFrame({
    "product_id": range(1001, 1021),
    "product_name": [f"Product_{i}" for i in range(1, 21)],
    "current_price": np.random.randint(200, 2000, 20)
})

products.head()
```

Creating Sales Dataset

```python
sales = pd.DataFrame({
    "product_id": np.random.choice(range(1001, 1021), 300),
    "quantity_sold": np.random.randint(1, 6, 300),
    "transaction_date": pd.date_range(
        end=pd.Timestamp.today(), periods=300
    )
})

sales.head()
```

Creating Inventory Dataset

```python
inventory = pd.DataFrame({
    "product_id": range(1001, 1021),
    "inventory_on_hand": np.random.randint(20, 500, 20)
})

inventory.head()
```

Creating Competitor Prices

```python
competitor_prices = pd.DataFrame({
    "product_id": range(1001, 1021),
    "competitor_price": np.random.randint(180, 1900, 20)
})

competitor_prices.head()
```

Saving all raw data

```python
products.to_csv(os.path.join(RAW_DATA_PATH, "products.csv"), index=False)
sales.to_csv(os.path.join(RAW_DATA_PATH, "store_sales_line_items.csv"), index=False)
inventory.to_csv(os.path.join(RAW_DATA_PATH, "store_inventory_snapshot.csv"), index=False)
competitor_prices.to_csv(os.path.join(RAW_DATA_PATH, "competitor_prices.csv"), index=False)

print("✅ All CSV files saved successfully at:", RAW_DATA_PATH)
```
